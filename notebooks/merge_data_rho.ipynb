```python
import pickle as pkl
import hist
import numpy as np
ratio_data_mc = [0.76056319, 0.97830354 ,1.24357577,  1.20827736]
```

```python
outputs = [ pkl.load(open("./inputs/rho/rho_data_2016APV.pkl", "rb")),
            pkl.load(open("./inputs/rho/rho_data_2016.pkl", "rb")),
            pkl.load(open("./inputs/rho/rho_data_2017.pkl", "rb")),
            pkl.load(open("./inputs/rho/rho_data_2018.pkl", "rb"))
          ]


hist_keys = ['ptjet_rhojet_u_reco', 'ptjet_rhojet_g_reco']
out_dict = {}
for i, output in enumerate(outputs):
    for key in hist_keys:
        if i == 0:
            out_dict[key] = output[key].project('ptreco', 'mpt_reco', 'systematic')
        else:
            out_dict[key] += output[key].project('ptreco', 'mpt_reco', 'systematic')

pkl.dump(out_dict, open("./inputs/rho/rho_data_all.pkl", "wb"))
```

```python
##V2 MC

outputs = [ pkl.load(open("./inputs/rho/jms_pythiaV2_2016APV_syst.pkl", "rb")),
            pkl.load(open("./inputs/rho/jms_pythiaV2_2016_syst.pkl", "rb")),
            pkl.load(open("./inputs/rho/jms_pythiaV2_2017_syst.pkl", "rb")),
            pkl.load(open("./inputs/rho/jms_pythiaV2_2018_syst.pkl", "rb"))
          ]


hist_keys = ['ptjet_rhojet_u_reco', 'ptjet_rhojet_g_reco', 'response_matrix_rho_u', 'response_matrix_rho_g', 'ptjet_rhojet_u_gen', 'ptjet_rhojet_g_gen',]
out_dict = {}
for i, output in enumerate(outputs):
    for key in hist_keys:
        if key in ['ptjet_rhojet_u_reco', 'ptjet_rhojet_g_reco']:

            if i == 0:
                out_dict[key] = output[key].project('ptreco', 'mpt_reco', 'systematic')
            else:
                out_dict[key] += output[key].project('ptreco', 'mpt_reco', 'systematic')
            print(f"Processed {key} for file {i}, with sum {output[key].project('ptreco', 'mpt_reco', 'systematic').sum().value}")
            print("Current total sum:", out_dict[key].sum().value)
        elif key in ['response_matrix_rho_u', 'response_matrix_rho_g']:
            if i == 0:
                out_dict[key] = output[key].project('ptgen', 'mpt_gen', 'ptreco', 'mpt_reco', 'systematic')
            else:
                out_dict[key] += output[key].project('ptgen', 'mpt_gen', 'ptreco', 'mpt_reco', 'systematic')
            #print(f"Processed {key} for file {i}")
        elif key in ['ptjet_rhojet_u_gen', 'ptjet_rhojet_g_gen']:
            if i == 0:
                out_dict[key] = output[key].project('ptgen', 'mpt_gen', 'systematic')
            else:
                out_dict[key] += output[key].project('ptgen', 'mpt_gen', 'systematic')
            #print(f"Processed {key} for file {i}")
```
