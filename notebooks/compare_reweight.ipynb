```python
import pickle as pkl
import hist
import matplotlib.pyplot as plt

with open("./inputs/massInputs/mass_pythia_all.pkl", "rb") as handle:
    output_pythia = pkl.load(handle)

with open("./inputs/massInputs/mass_herwig_all.pkl", "rb") as handle:
    output_herwig = pkl.load(handle)

with open("./inputs/massInputs/mass_reweight_pythia_all.pkl", "rb") as handle:
    output_reweighted = pkl.load(handle)
```

```python
for i in range(4):
    h2d_h = output_herwig['ptjet_mjet_u_gen'].project("ptgen", "mgen")
    h2d_prw = output_reweighted['ptjet_mjet_u_gen'].project("ptgen", "mgen")
    h2d_p = output_pythia['ptjet_mjet_u_gen'].project("ptgen", "mgen", 'systematic')
    h2d_h[i, :hist.loc(190)].plot(density=True, label="Herwig", histtype="step", color="red")
    h2d_p[i, :hist.loc(190), 'nominal'].plot(density=True, label="Pythia", histtype="step", color="blue")
    h2d_prw[i, :hist.loc(190)].plot(density=True, label="Reweighted", histtype="step", color="green", ls = '--')
    plt.legend()
    plt.show()
```
